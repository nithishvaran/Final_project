```python
import pandas as pd
import os
import librosa

audio_dataset_path='C:/Users/NITHISHVARAN T P/development'
metadata=pd.read_csv('C:/Users/NITHISHVARAN T P/development.csv')
metadata
```

```python
import numpy as np
from tqdm import tqdm
extracted_features=[]
for index_num,row in tqdm(metadata.iterrows()):
    file_name = os.path.join(os.path.abspath(audio_dataset_path),str(row["file_name"]))
    final_class_labels=row["caption_1"]
    data=features_extractor(file_name)
    extracted_features.append([data,final_class_labels])
```

```python
extracted_features_df=pd.DataFrame(extracted_features,columns=['features','captions'])
extracted_features_df.head()
#extracted_features_df.to_csv('file1.csv')
```

```python
### Split the dataset into independent and dependent dataset
X=np.array(extracted_features_df['features'].tolist())
y=np.array(extracted_features_df['captions'].tolist())
```

```python
X.shape
```

```python
### Label Encoding
y=np.array(pd.get_dummies(y))
```

```python
### Train Test Split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

# Model Creation

```python
import tensorflow as tf
print(tf.__version__)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Dropout,Activation,Flatten
from tensorflow.keras.optimizers import Adam
from sklearn import metrics
```

```python
### No of classes
num_labels=y.shape[1]
```

```python
model=Sequential()
###first layer
model.add(Dense(100,input_shape=(13,)))
model.add(Activation('relu'))
model.add(Dropout(0.5))
###second layer
model.add(Dense(200))
model.add(Activation('relu'))
model.add(Dropout(0.5))
###third layer
model.add(Dense(100))
model.add(Activation('relu'))
model.add(Dropout(0.5))
###final layer
model.add(Dense(num_labels))
model.add(Activation('softmax'))
```

```python
model.summary()
```

```python
model.compile(loss='categorical_crossentropy',metrics=['accuracy'],optimizer='adam')
```

```python
## Trianing my model
from tensorflow.keras.callbacks import ModelCheckpoint
from datetime import datetime 

num_epochs = 100
num_batch_size = 32

checkpointer = ModelCheckpoint(filepath='saved_models/audio_classification.hdf5', 
                               verbose=1, save_best_only=True)
start = datetime.now()

model.fit(X_train, y_train, batch_size=num_batch_size, epochs=num_epochs, validation_data=(X_test, y_test), callbacks=[checkpointer], verbose=1)

duration = datetime.now() - start
print("Training completed in time: ", duration)
```

```python
test_accuracy=model.evaluate(X_test,y_test,verbose=0)
print(test_accuracy[1])
```
